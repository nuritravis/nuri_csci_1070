```python
import pandas as pd

belize_df = pd.read_csv("belize_airbnb_listings.csv")

grouped_df = belize_df.groupby('neighbourhood')

# average price
price_df = grouped_df[["price"]].mean()
price_df
```

```python
# total number of reviews
reviews_df = grouped_df[["number_of_reviews"]].count()
reviews_df
```

```python
# max availability
availability_df = grouped_df[["availability_365"]].max()
availability_df
availability_df
```

```python
merged_df_1 = pd.merge(availability_df, price_df, on="neighbourhood")
merged_df_2 = pd.merge(merged_df_1, reviews_df, on="neighbourhood")
merged_df_2
```
